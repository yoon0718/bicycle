```python
#ADDR1 : 거치대주소
#STATION_NM : 대여소이름
#LOCKER : 거치대개수
#Y_POS : 위도
#X_POS : 경도
for i in range(len(data)):
    name = data.loc[i, 'STATION_NM']
    name2 = data.loc[i, 'ADDR1']
    total = data.loc[i, 'LOCKER']
    lat = data.loc[i, 'Y_POS']
    long = data.loc[i, 'X_POS']
    print(name, name2, total, lat, long)
```

```python
for i in range(len(data)):
    name = data.loc[i, 'STATION_NM']
    name2 = data.loc[i, 'ADDR1']
    total = data.loc[i, 'LOCKER']
    lat = data.loc[i, 'Y_POS']
    long = data.loc[i, 'X_POS']
    icon=folium.Icon(color='green', icon='info-sign')
    folium.Marker(location = [lat, long],
                 tooltip = f"{name} : {total}개 {name2}",
                 icon = icon
                 ).add_to(m)
m
```
