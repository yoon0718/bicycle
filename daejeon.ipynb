```python
data = pd.read_excel('C:/Users\math-185/tashubicycle.xlsx')
data.head()
#STS_NM : 대여소이름 name
#LOCK_ON : 자전거주차총건수 available
#TOT_CNT : 거치대개수 total
#Y_POS : 위도 lat
#X_POS : 경도 long
```

```python
import folium
```

```python
m = folium.Map(location = ['36.34034903914251', '127.3921756688959'], zoom_start = 13)
m
```

```python
#STS_NM : 대여소이름 name
#LOCK_ON : 자전거주차총건수 available
#TOT_CNT : 거치대개수 total
#Y_POS : 위도 lat
#X_POS : 경도 long
for i in range(len(data)):
    name = data.loc[i, 'STS_NM']
    available = data.loc[i, 'LOCK_ON']
    total = data.loc[i, 'TOT_CNT']
    lat = data.loc[i, 'Y_POS']
    long = data.loc[i, 'X_POS']
    if available/total > 0.5:
        color = 'blue'
    elif available < 2:
        color = 'red'
    else:
        color = 'green'
    icon=folium.Icon(color=color, icon='info-sign')
    folium.Marker(location = [lat, long],
                 tooltip = f"{name} : {available}", 
                  icon = icon
             ).add_to(m)
m
```

```python
from folium.plugins import MiniMap, MarkerCluster
```

```python
m2 = folium.Map(location = ['36.34034903914251', '127.3921756688959'], zoom_start = 13)
#미니맵 추가
minimap = MiniMap()
m2.add_child(minimap)
#클러스터
marker_cluster2 = MarkerCluster().add_to(m2)

for i in range(len(data)):
    name = data.loc[i, 'STS_NM']
    available = data.loc[i, 'LOCK_ON']
    total = data.loc[i, 'TOT_CNT']
    lat = data.loc[i, 'Y_POS']
    long = data.loc[i, 'X_POS']
    #마커 추가
    #자전거 수량에 대해 색상으로 표시
    #자전거 보유율 50% 초과 > 파란색
    #현재 자전거 2대보다 적게 있을 경우 > 빨간색
    #현재 자전거 2대보다 많고 자전거 보유율 50% 미만 > 초록
    if available/total > 0.5:
        color = 'blue'
    elif available < 2:
        color = 'red'
    else:
        color = 'green'
    icon=folium.Icon(color=color, icon='info-sign')
    folium.Marker(location = [lat, long],
                 tooltip = f"{name} : {available}", 
                  icon = icon
             ).add_to(marker_cluster2)
m2
```
